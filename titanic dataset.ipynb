```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
training_data = pd.read_csv("train.csv")
```

```python
training_data
```

```python
training_data.describe()
```

```python
## The best possible Age to fill in place of nan is the mean that is 30
training_data["Age"].fillna(training_data["Age"].median(), inplace = True)
training_data.describe()
```

```python
training_data.columns
```

```python
## Let us drop cabin and name as cabin is filled with na and name seems to have no significance.
training_data.drop("Name", axis = 1, inplace = True)
training_data
```

```python
## On Deeper analysis of the Cabin Column we found that the cabin is A to F 
## So filling na with 0, and A to F with respective integers 1 to 6
training_data["Cabin"].fillna(0, inplace = True)
training_data

```

```python
def Allot(obj) :
    if obj == 0 :
        return int(obj)
    char = obj[0]
    return char

training_data["Cabin"] = training_data["Cabin"].apply(Allot)
training_data
```

```python
training_data["Cabin"][6]
```

```python
training_data["Cabin_C"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "C" :
        return 1
    else :
        return 0

training_data["Cabin_C"] = training_data["Cabin_C"].apply(Allot)
training_data
```

```python
training_data["Cabin_C"][6]
```

```python
training_data["Cabin_B"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "B" :
        return 1
    else :
        return 0

training_data["Cabin_B"] = training_data["Cabin_B"].apply(Allot)
training_data
```

```python
training_data["Cabin_A"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "A" :
        return 1
    else :
        return 0

training_data["Cabin_A"] = training_data["Cabin_A"].apply(Allot)
training_data
```

```python
training_data["Cabin_A"][6]
```

```python
## Sex is male or female, let us replace them with 0 and 1 respectively
training_data["Sex"].replace("female", 1, inplace = True)
training_data["Sex"].replace("male", 0, inplace = True)
training_data
```

```python
def convert(obj) :
    if obj == 0 :
        return 1
    
    else :
        return 0

training_data["Sex_M"] = training_data["Sex"].apply(convert)

training_data
```

```python
## Embarked is S or Q or C, so S for 0 and Q for 1 and C for 2 
def Convert(obj) :
    if obj == "S" :
        return 1
    else : 
        return 0

training_data["Embarked_S"] = training_data["Embarked"].apply(Convert)
```

```python
## Embarked is S or Q or C, so S for 0 and Q for 1 and C for 2 
def Convert(obj) :
    if obj == "Q" :
        return 1
    else : 
        return 0

training_data["Embarked_Q"] = training_data["Embarked"].apply(Convert)
```

```python
## Embarked is S or Q or C, so S for 0 and Q for 1 and C for 2 
def Convert(obj) :
    if obj == "C" :
        return 1
    else : 
        return 0

training_data["Embarked_C"] = training_data["Embarked"].apply(Convert)
training_data
```

```python
training_data.drop("Embarked", axis = 1, inplace = True)
training_data
```

```python
training_data.drop("Ticket", axis = 1, inplace = True)
training_data
```

```python
Y = np.array(training_data["Survived"])
Y
```

```python
training_data.drop("Survived", axis = 1, inplace = True)
training_data
```

```python
training_data["Embarked_S"].fillna(0, inplace = True)
training_data
```

```python
training_data["Embarked_Q"].fillna(0, inplace = True)
training_data["Embarked_C"].fillna(0, inplace = True)
training_data
```

```python
training_data["Cabin_D"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "D" :
        return 1
    else :
        return 0

training_data["Cabin_D"] = training_data["Cabin_D"].apply(Allot)
training_data
```

```python
training_data["Cabin_E"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "E" :
        return 1
    else :
        return 0

training_data["Cabin_E"] = training_data["Cabin_E"].apply(Allot)
training_data
```

```python
training_data["Cabin_F"] = training_data["Cabin"]
def Allot(obj) :
    if obj == 0 :
        return 0
    
    if obj == "F" :
        return 1
    else :
        return 0

training_data["Cabin_F"] = training_data["Cabin_F"].apply(Allot)
training_data
```

```python
training_data.drop("Cabin", inplace = True, axis = 1)
training_data
```

```python
X = training_data.to_numpy()
X
```

```python
Log = LogisticRegression(max_iter = 500)
Log.fit(X, Y)
```

```python
Log.score(X, Y)
```
